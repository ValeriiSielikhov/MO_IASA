```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
from matplotlib import cm

eps = 1e-5
```

```python
def f(x):
    #return 2*x[0]**2 + 8*x[1]**2 + 3*x[2]**2+0.001*x[0]*x[2]-x[0]-x[1]
    return 100*(x[1] - x[0]**3)**2 + 100*(1 - x[0])**2
```

```python
def grad(f, x, h=1e-3):
    if x.shape[0] == 2:
        x_grad = (f([x[0] + h, x[1]]) - f([x[0] - h, x[1]])) / (2*h)
        y_grad = (f([x[0], x[1] + h]) - f([x[0], x[1] - h])) / (2*h)
        return np.array([x_grad, y_grad], dtype=float)
    elif x.shape[0] ==3:
        x_grad = (f([x[0] + h, x[1], x[2]]) - f([x[0] - h, x[1], x[2]])) / (2*h)
        y_grad = (f([x[0], x[1] + h, x[2]]) - f([x[0], x[1] - h, x[2]])) / (2*h)
        z_grad = (f([x[0], x[1], x[2] +h]) - f([x[0], x[1], x[2]-h])) / (2*h)
        return np.array([x_grad, y_grad, z_grad], dtype=float)
    else:
        print("error")
```

```python
def second_method(f, p):
    it = 0
    h = np.array([-grad(f, p[0])])
    alpha = minimize_scalar(lambda l: f(p[-1] + l * h[-1])).x
    p = np.append(p, [p[-1] + alpha * h[-1]], axis=0)
    b = np.array([])
    while np.linalg.norm(p[-2]-p[-1]) > eps:
        it+= 1
        print('it:', it, 'point:', p[-1],'value:', f(p[-1]) )
        if it%p.shape[1] == 0:
            b = np.append(b, 0.)
        else:
            b = np.append(b, grad(f, p[-1])@(grad(f, p[-1]) - grad(f, p[-2]))/np.linalg.norm(grad(f, p[-2]))**2)
        h = np.append(h, [-grad(f, p[-1]) + b[-1]*h[-1]], axis=0)
        alpha = minimize_scalar(lambda l: f(p[-1] + l * h[-1])).x
        p = np.append(p, [p[-1] + alpha * h[-1]], axis=0)
        
    return p
```

```python
#p = np.array([[10,10, 10]])
p = np.array([[10,10]])
new_p = second_method(f, p)
```

```python
def makeGrid(range, step):
    x = np.arange(-range, range, step)
    y = np.arange(-range, range, step)
    xgrid, ygrid = np.meshgrid(x, y)
    zgrid = f([xgrid, ygrid])
    return xgrid, ygrid, zgrid

xt, yt, zt = makeGrid(9, 0.1)
fig = plt.figure(figsize=(18, 9))

ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(xt, yt, zt, color = 'white', )
ax.plot(new_p.T[0], new_p.T[1], f(new_p.T),  color = "green", marker = "o")
plt.show()
```
